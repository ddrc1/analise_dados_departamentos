```python
creditcard_df.drop(columns=["CUST_ID"], inplace=True)
creditcard_df.head()
```

```python
creditcard_df.isna().sum()
```

```python
creditcard_df.loc[creditcard_df.MINIMUM_PAYMENTS.isna(), "MINIMUM_PAYMENTS"] = creditcard_df.MINIMUM_PAYMENTS.mean()
creditcard_df.loc[creditcard_df.CREDIT_LIMIT.isna(), "CREDIT_LIMIT"] = creditcard_df.CREDIT_LIMIT.mean()

creditcard_df.isna().sum()
```

```python
creditcard_df.duplicated().sum()
```

```python
plt.figure(figsize=(10, 50))
for i, column in enumerate(creditcard_df.columns):
    plt.subplot(17, 1, i+1)
    plt.hist(creditcard_df[column], bins=50)
    plt.title(column)
plt.tight_layout()
```

```python
correlations = creditcard_df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlations, annot=True);
```
