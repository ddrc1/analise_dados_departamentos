## Definição de grupos de clientes utilizando K-Means

Normalizando o dado por min max

```python
scaler = MinMaxScaler()
creditcard_df_scaled = scaler.fit_transform(creditcard_df)
```

#### Definição do número de clusters usando o Elbow Method

```python
wcss1 = []
clusters = range(1, 21)
for i in clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(creditcard_df_scaled)
    wcss1.append(kmeans.inertia_)
    
wcss1
```

```python
plt.figure(figsize=(10, 6))
plt.plot(clusters, wcss1, "o--")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.xticks(range(1, 21, 2));
```

```python
abs(np.diff(np.array(wcss1)))
```

#### Agrupamento com o K-Means

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(creditcard_df_scaled)
labels = kmeans.labels_

labels
```

Quantidade de clientes por grupo

```python
np.unique(labels, return_counts=True)
```

Valores utilizando o centroide de cada cluster

```python
centers = kmeans.cluster_centers_
centers = scaler.inverse_transform(centers)
cluster_centers = pd.DataFrame(centers, columns=creditcard_df.columns)

cluster_centers
```

#### Análise de alguns grupos
Grupo 2: São clientes mais novos. Não fazem tantas compras, sacam mais dinheiro e possuem um limite de crédito mais baixo.

Grupo 3: São os clientes mais rentáveis. Mantém mais dinheiro em conta, possuem maior movimentação de dinheiro na conta. Compram muito, tendendo a pagar tanto à vista quanto à prazo e costumam atrasar parcelas, gerando mais lucro pro banco por meio de juros. Pode-se dizer que são clientes VIP.

Grupo 4: Possuem uma quantidade razoável de dinheiro em conta, tendem a fazer mais compras parceladas, possuem um alto limite de crédito têm alto valor e tendem a pagar grandes parcelas e costumam atrasar seu pagamento.

Grupo 6: não possuem muito dinheiro no saldo, mas são os clientes que mais compram, utilizando com mais frequencia o cartão, usando-o mais para compras à vista. Das vezes que compram parcelado, tendem não atrasar

Grupo 8: São os clientes que usam muito pouco a conta bancária. Têm pouco dinheiro em conta, quase não movimentam dinheiro em conta, realizam poucas compras.

```python
creditcard_df_cluster = creditcard_df.copy()
creditcard_df_cluster["cluster"] = labels

creditcard_df_cluster
```

#### Visualização dos clusters

```python
for col in creditcard_df.columns:
    plt.figure(figsize=(35, 5))
    for j in range(10):
        plt.subplot(1, 11, j+1)
        cluster = creditcard_df_cluster.query(f"cluster == {j}")
        cluster[col].hist(bins=20)
        plt.title(f"{col} \nCluster {j}")
    plt.show()
```

## Redução de dimensionalidade utilizando Auto Encoders <br>
A ideia principal é reduzir a dimensionalidade do dado (17 -> 10) para que a clusterização do K-Means fique mais precisa

```python
input_layer = layers.Input(shape=creditcard_df_scaled.shape[1])
x = layers.Dense(500, "relu")(input_layer)
x = layers.Dense(2000, "relu")(x)
encoded = layers.Dense(10, "relu")(x)
x = layers.Dense(2000, "relu")(encoded)
x = layers.Dense(500, "relu")(x)
decoded = layers.Dense(creditcard_df_scaled.shape[1], "relu")(x)

autoencoder = models.Model(input_layer, decoded)
encoder = models.Model(input_layer, encoded)

optimizer = optimizers.Adam(learning_rate=0.001)
autoencoder.compile(optimizer=optimizer, loss="mse")
autoencoder.summary()
```

```python
history = autoencoder.fit(creditcard_df_scaled, creditcard_df_scaled, epochs=40)
```

```python
compact = encoder.predict(creditcard_df_scaled)
compact.shape
```

### Definição de grupos de clientes utilizando K-Means
#### Definição do número de clusters usando o Elbow Method

```python
wcss2 = []
clusters = range(1, 21)
for i in clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(compact)
    wcss2.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(clusters, wcss1, "o--", label="Without Auto Encoder")
plt.plot(clusters, wcss2, "o--", label="With Auto Encoder")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.xticks(range(1, 21, 2))
plt.legend();
```

```python
abs(np.diff(np.array(wcss2)))
```

#### Agrupamento com o K-Means

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(compact)

labels = kmeans.labels_
creditcard_df['cluster'] = labels
creditcard_df
```
